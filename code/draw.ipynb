```python
from matplotlib import pyplot as plt
```

```python
loss_path = '../loss_pre.txt'
loss_pre = []
for line in open(loss_path, encoding='utf-8'):
    loss_pre.append(float(line.split(':')[-1].strip(' ')))
```

```python
loss_after_path = '../loss_after.txt'
loss_after = []
for line in open(loss_after_path, encoding='utf-8'):
    loss_after.append(float(line.split(':')[-1].strip(' ')))
```

```python
plt.plot(loss_pre, color='b', label='before_aug')
plt.plot(loss_after, color='g', label='after_aug')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(loc=1)
plt.show()
#标签展示位置，数字代表标签具位置
```
